## Завдання
Створити клас ProductCalculator, який:
- Приймає в конструкторі список чисел
- Має метод calculate_product(), який обчислює добуток всіх елементів списку
- Має property product, яка повертає добуток всіх чисел

```python
class ProductCalculator:
    def __init__(self, numbers: list[int]):
        self._numbers = numbers
        self._product = None

    def calculate_product(self) -> int:
        result = 1
        for number in self._numbers:
            result *= number
        self._product = result
        return result

    @property
    def product(self) -> int:
        if self._product is None:
            self.calculate_product()
        return self._product

    def __str__(self) -> str:
        return "Ім'я студента: Богдан, Група: КН-210"


data_sets = [
    [1, 2, 3, 4],
    [2, 5, 10],
    [3, 3, 3]
]

calculators = []

for data in data_sets:
    calculator = ProductCalculator(data)
    calculators.append(calculator)

for calculator in calculators:
    print(str(calculator))
    print(f"Вхідні дані: {calculator._numbers}")
    print(f"Добуток елементів списку: {calculator.product}")
    print("-" * 40)
```
