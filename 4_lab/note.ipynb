# Вчимось прауцювати з ООП у Python
### Завдання:
- створити програму імітації роботи бібліотеки з використанням об'єктно-орієнтованого програмування (ООП) у Python.
- програма повинна складатись з декількох класів, які представляють Бібліокеку, Книгу та Студента. Кожен клас повинен знаходитись у власному файлі. Головна програма імітації роботи бібліотеки повинна бути у файлі main.py.
- клас Бібліотека повинен мати методи для додавання книг, пошуку книг за назвою та автором, а також для видачі книг студентам та повернення книг.
- клас Книга повинен містити інформацію про назву, автора, рік видання та статус (доступна чи видана). Початково всі книги повинні бути доступні.
- клас Студент повинен мати інформацію про студента (імя та прізвище), мати властивість яка описує студента та містити список книг, які він взяв.
- у файлі main.py початково повинні бути створені кілька об'єктів класу Книга та додані до об'єкту класу Бібліотека. Також повинні бути створені об'єкти класу Студент, які можуть брати та повертати книги з бібліотеки.
